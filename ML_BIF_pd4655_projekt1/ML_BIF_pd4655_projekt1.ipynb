```python
import pandas as pd

df = pd.read_csv("dane_projekt1.csv")
print(df.head())
print(df.info())
```

```python
#Eksploracja danych
#W zbiorze są cztery klasy strucutral protein - najwiecej przykładw, receptor, enzyme i transcription_factor
#Zbiór nie jest mocno niezbalansowany co sprzyjac moze trenowaniu modelu ML
#Dane sa normalnie rozproszone, ale mogą zaweirac wartosci skrajne 
```

```python
#Przetwarzanie danych
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gene_Function'] = le.fit_transform(df['Gene_Function'])
```

```python
#Podział na zbiory treningowe i testowe
from sklearn.model_selection import train_test_split

X = df.iloc[:, 1:11]  # Tissue_1 to Tissue_10
y = df['Gene_Function']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
#Budowa modelu ML
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
#Ewaluacja modelu i interpretacja wyników
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
